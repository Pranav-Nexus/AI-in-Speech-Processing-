```python
import librosa
import matplotlib.pyplot as plt
from IPython.display import Audio

# Load the audio file
filename = 'AI in Speech Processing.wav' 
y, sr = librosa.load(filename)

# Calculate the time axis
time = librosa.times_like(y, sr=sr)

# Plot the waveform
plt.figure(figsize=(14, 5))
plt.plot(time, y[::-1])  # reverse the y array
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Waveform')
plt.show()

# Function to play a segment of the audio
def play_segment(start_sec, end_sec):
    # Ensure that start_sec is less than end_sec
    if start_sec > end_sec:
        start_sec, end_sec = end_sec, start_sec  # swap the two values

    start_sample = int(start_sec * sr)
    end_sample = int(end_sec * sr)
    return Audio(data=y[start_sample:end_sample][::-1], rate=sr)  # reverse the audio segment

# Play different segments
play_segment(5, 0)  # play the first 5 seconds in reverse
```
